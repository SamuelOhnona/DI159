# 🌿 Exercise 2: Identify Unstocked Plants Using openpyxl

This exercise loads an Excel workbook listing plants and checks which ones are **not in stock** (indicated by "No" in column H).  
The program starts at cell A2 and continues until it hits a blank cell in column A.

Excel file path used:
`/Users/sam/Documents/DI_DATA_2025/week12/Plants.xlsx`


```python
from openpyxl import load_workbook

# Define the file path
file_path = "/Users/sam/Documents/DI_DATA_2025/week12/Plants.xlsx"

# Load the workbook and the first sheet
workbook = load_workbook(file_path)
sheet = workbook["Sheet1"]

# Start from cell A2
current_cell = sheet["A2"]

# Loop through rows until an empty A cell is found
print("Unstocked plants:")
while current_cell.value is not None:
    # Get the corresponding H cell (7 columns to the right)
    stock_cell = sheet.cell(row=current_cell.row, column=current_cell.column + 7)
    
    if stock_cell.value == "No":
        print("-", current_cell.value)
    
    # Move to the next row (same column A)
    current_cell = sheet.cell(row=current_cell.row + 1, column=1)
```

```python
# You can optionally add a safety limit to prevent infinite loops
# Uncomment this block and use instead of the loop above if needed

print("Unstocked plants (with safety limit):")
max_rows = 1000
count = 0
while current_cell.value is not None and count < max_rows:
    stock_cell = sheet.cell(row=current_cell.row, column=current_cell.column + 7)
    if stock_cell.value == "No":
        print("-", current_cell.value)
    current_cell = sheet.cell(row=current_cell.row + 1, column=1)
    count += 1
```
