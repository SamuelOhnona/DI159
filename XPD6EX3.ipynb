# 📊 Exercise 3: Filter Excel Data with Pandas and Write with openpyxl

**Goal:**  
- Load an Excel file using `pandas`  
- Filter the rows where **Sales > 1000**  
- Write the filtered data back into the same Excel file using `openpyxl`

**File path:**  
`/Users/sam/Documents/DI_DATA_2025/week12/data.xlsx`

```python
import pandas as pd
from openpyxl import load_workbook
```

```python
# Define file path
file_path = "/Users/sam/Documents/DI_DATA_2025/week12/data.xlsx"

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(file_path)

# Preview the data
df.head()
```

```python
# Filter rows where 'Sales' is greater than 1000
filtered_df = df[df["Sales"] > 1000]

# Preview filtered data
filtered_df.head()
```

```python
# Load the original Excel workbook
book = load_workbook(file_path)

# Create an ExcelWriter object with openpyxl engine
with pd.ExcelWriter(file_path, engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
    # Write filtered data into a new sheet called "Filtered Sales"
    filtered_df.to_excel(writer, sheet_name="Filtered Sales", index=False)

print("✅ Filtered data written to 'Filtered Sales' sheet.")
```
