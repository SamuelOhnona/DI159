# 🧮 Exercise 1: Create an Excel Calculator using openpyxl

This exercise uses the `openpyxl` library to generate an Excel file with:
- A first number in cell B1
- A second number in cell B2
- A formula in cell B3 to multiply them

Expected result: the Excel file displays `16` in B3.

```python
from openpyxl import Workbook

# Create a new workbook and select the active sheet
wb = Workbook()
ws = wb.active

# Add labels and values
ws['A1'] = "First number ==> "
ws['B1'] = 2

ws['A2'] = "Second number ==> "
ws['B2'] = 8

# Add a formula to multiply B1 and B2
ws['B3'] = "=B1*B2"

# Save the workbook
filename = "XPD6EX1_calculator.xlsx"
wb.save(filename)

print(f"Workbook '{filename}' created successfully.")
```

```python
import os

if os.path.exists(filename):
    print("✅ File created and saved correctly.")
else:
    print("❌ File was not created.")
```
