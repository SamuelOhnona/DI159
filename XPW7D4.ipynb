```python
# 🌟 Exercise 2 & 3: Load Datasets and Classify Data Types
# ----------------------------------------------------------

# Update these paths to match your extracted .csv files
sleep_df = pd.read_csv("How Much Sleep Do Americans.csv")
mental_df = pd.read_csv("Global Trends in Mental Health Disorder.csv")
credit_df = pd.read_csv("Credit Card Approvals (Clean Data).csv")

# Show previews
display(sleep_df.head())
display(mental_df.head())
display(credit_df.head())

# Example classification for credit_df
credit_df.dtypes  # Quantitative vs Qualitative to be discussed below

# Markdown cell for classification
'''
## Exercise 3: Data Type Classification

### Credit Card Approvals:
- `Age`: Quantitative (numeric, continuous)
- `Gender`: Qualitative (categorical)
- `Income`: Quantitative
- `Approval`: Qualitative (Yes/No)

### Sleep Dataset:
- `State`: Qualitative
- `Hours Sleep`: Quantitative

### Mental Health Dataset:
- `Country`: Qualitative
- `Disorder Rate`: Quantitative
- `Year`: Quantitative (but sometimes used categorically)
'''
```

```python
# 🌟 Exercise 4: Iris Dataset
# --------------------------------------

from sklearn.datasets import load_iris
iris = load_iris(as_frame=True)
iris_df = iris.frame

# Preview
iris_df.head()

# Markdown cell for classification
'''
## Iris Dataset Data Types
- `sepal length (cm)`: Quantitative
- `sepal width (cm)`: Quantitative
- `petal length (cm)`: Quantitative
- `petal width (cm)`: Quantitative
- `target`: Qualitative (species label)
'''
```

```python
# 🌟 Exercise 5: Basic Analysis & Plotting
# --------------------------------------

# Mean, median, mode of petal length
print("Mean:", iris_df['petal length (cm)'].mean())
print("Median:", iris_df['petal length (cm)'].median())
print("Mode:", iris_df['petal length (cm)'].mode()[0])

# Histogram
plt.hist(iris_df['petal length (cm)'], bins=20, color='skyblue')
plt.title('Petal Length Distribution')
plt.xlabel('Length (cm)')
plt.ylabel('Frequency')
plt.show()
```
