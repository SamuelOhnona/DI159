Cellule 1 – Imports & Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Optional: for inline plotting
%matplotlib inline
```

Cellule 2 – Data Preparation

```python
# Define cities and months
cities = ['Paris', 'New York', 'Tokyo', 'Sydney', 'Moscow', 'Cape Town', 'Rio', 'Toronto', 'Mumbai', 'Dubai']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Generate synthetic temperature data (°C)
temp_data = np.random.uniform(low=-5, high=35, size=(10, 12)).round(1)

# Create DataFrame
df_temp = pd.DataFrame(temp_data, index=cities, columns=months)

# Display first few rows
df_temp.head()
```

Cellule 3 – Data Analysis

```python
# Calculate annual average temperature per city
df_temp['Annual Avg'] = df_temp.mean(axis=1).round(2)

# Identify city with highest and lowest average temperature
hottest_city = df_temp['Annual Avg'].idxmax()
coldest_city = df_temp['Annual Avg'].idxmin()

print(f"Hottest city: {hottest_city} ({df_temp.loc[hottest_city, 'Annual Avg']}°C)")
print(f"Coldest city: {coldest_city} ({df_temp.loc[coldest_city, 'Annual Avg']}°C)")
```

Cellule 4 – Data Visualization: Line Plot (monthly trend per city)

```python
plt.figure(figsize=(14, 7))
for city in cities:
    plt.plot(months, df_temp.loc[city, months], label=city)

plt.title('Monthly Average Temperature per City')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

Cellule 5 – Data Visualization: Bar Plot of Annual Avg Temperature

```python
plt.figure(figsize=(12, 6))
df_temp['Annual Avg'].sort_values().plot(kind='barh', color='skyblue')
plt.title('Annual Average Temperature by City')
plt.xlabel('Average Temperature (°C)')
plt.grid(True)
plt.show()
```
