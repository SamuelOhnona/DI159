# 📈 Exercise 4: Group Sales by Product and Export to Excel

This notebook reads sales data from an Excel file, groups it by product, creates a bar chart of total sales per product, and exports the result to a new Excel file named `sales_report.xlsx`.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path to the source Excel file
input_path = "/Users/sam/Documents/DI_DATA_2025/week12/productSales.xlsx"

# Read the Excel file into a DataFrame
df = pd.read_excel(input_path)

# Preview the data
df.head()
```

```python
# Group sales by product and calculate the total sales per product
grouped_df = df.groupby("product")["sales"].sum().reset_index()

# Show the result
grouped_df
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(grouped_df["product"], grouped_df["sales"], edgecolor='black')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Path to the output Excel file
output_path = "/Users/sam/Documents/DI_DATA_2025/week12/sales_report.xlsx"

# Export the grouped data to a new Excel file
with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    grouped_df.to_excel(writer, index=False, sheet_name="Summary")

print(f"✅ Sales report successfully written to: {output_path}")
```
