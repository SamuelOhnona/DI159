### 🟢 Exercise 1: Identifying Data Types

Below are various data sources. Identify whether each one is an example of structured or unstructured data.

1. **A company’s financial reports stored in an Excel file**  
   ➤ Structured Data – The information is organized in rows and columns, making it easy to query and analyze.

2. **Photographs uploaded to a social media platform**  
   ➤ Unstructured Data – Images do not follow a predefined model and require special processing to extract information.

3. **A collection of news articles on a website**  
   ➤ Unstructured Data – Although written in natural language, news articles can be analyzed using text mining or NLP.

4. **Inventory data in a relational database**  
   ➤ Structured Data – Stored in table formats with defined fields, it can be queried using SQL.

5. **Recorded interviews from a market research study**  
   ➤ Unstructured Data – Audio files need transcription and further processing to extract structured insights.
